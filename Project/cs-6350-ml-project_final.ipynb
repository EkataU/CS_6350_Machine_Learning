### Data Preprocess

```python
## Libraries
from sklearn import preprocessing
from sklearn.preprocessing import LabelBinarizer, LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline, Pipeline
```

```python
#Since it has 0 correlation, it can be dropped

df_tr.drop(['fnlwgt'], axis = 1, inplace = True)
df_ts.drop(['fnlwgt'], axis = 1, inplace = True)
print('identify columns:',df_tr.columns)
```

```python
#label encoder

#X_val = X_val.apply(LabelEncoder().fit_transform)
#X_train = X_train.apply(LabelEncoder().fit_transform)
#X_test = df_ts.apply(LabelEncoder().fit_transform)
```

```python
std_transformers = {}
print(f'Transformed columns in training dataset:............')
for col_name in ['age','education.num','capital.gain','capital.loss','hours.per.week']:
    std_transformers[col_name] = preprocessing.StandardScaler()
    
    df_tr[col_name] = std_transformers[col_name].fit_transform(np.array(df_tr[col_name])[:,np.newaxis])
    df_ts[col_name]  = std_transformers[col_name].transform(np.array(df_ts[col_name])[:,np.newaxis])
    
    print(col_name,' shape:',df_tr[col_name].shape)
```

```python
ord_transformers = {}
for col_name in ['workclass', 'education',  'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native']:
    ord_transformers[col_name] = preprocessing.OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    df_tr[col_name] = ord_transformers[col_name].fit_transform(np.array(df_tr[col_name])[:,np.newaxis])
    df_ts[col_name]  = ord_transformers[col_name].transform(np.array(df_ts[col_name])[:,np.newaxis])
```

```python
## Split the data into feats and labels
#train, valdiation and test dataset split
X,y = df_tr.drop(['income'],axis =1), df_tr.income
print('identify columns before dropping ID:',df_ts.columns)
X_test = df_ts.drop(columns ='ID')
print('identify columns after dropping ID:',X_test.columns)

#X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=44)
#y_train = LabelBinarizer().fit_transform(y_train)

# print('training dataset:',X_train.shape)
# print('training dataset:',type(y_train))
# print('validation dataset:',X_val.shape)
# print('test dataset:',df_ts.shape)
```

```python
# Multivariate imputer that estimates each feature
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp_mean  = IterativeImputer(max_iter=10, random_state=0)
X_train = pd.DataFrame(imp_mean.fit_transform(X), columns=X.columns)
X_test = pd.DataFrame(imp_mean.transform(X_test), columns=X_test.columns) 
```

```python
X_tr_imp, X_ts_imp = {}, {}
for col in X_train.columns:
    X_tr_imp[col] = X_train[col]
X_tr_imp['income'] = np.array(df_tr.income)

X_ts_imp['ID'] = np.array(df_ts.ID)
for col in X_test.columns:
    X_ts_imp[col] = X_test[col]

X_tr_imp = pd.DataFrame(X_tr_imp)
X_ts_imp = pd.DataFrame(X_ts_imp)
```

```python
for col_name in ['workclass', 'occupation', 'native']:
  
    X_tr_imp[col_name] = round(X_tr_imp[col_name])
 
    X_ts_imp[col_name]  = round(X_ts_imp[col_name])
   
```

```python
## Make a deep copy
X_tr_inv = X_tr_imp.copy()
X_ts_inv  = X_ts_imp.copy()
```

```python
## Reverse the ordered mapping
for col_name, transformer in ord_transformers.items():
    X_tr_inv[col_name] = transformer.inverse_transform(np.array(X_tr_inv[col_name])[:,np.newaxis])
    X_ts_inv[col_name]  = transformer.inverse_transform(np.array(X_ts_inv[col_name])[:,np.newaxis])
```

```python
## Split the data into feats and labels
X_tr_inv_feats, X_tr_inv_labels = X_tr_inv.drop(columns='income'), X_tr_inv.income
X_ts_inv_feats             = X_ts_inv
```

```python
## Take care of non_numeric features
categories = [
    ## workclass
    ['Private', 
     'Self-emp-not-inc', 
     'Self-emp-inc', 
     'Federal-gov', 
     'Local-gov', 
     'State-gov', 
     'Without-pay', 
     'Never-worked'],
    ## education
    ['Bachelors', 
     'Some-college', 
     '11th', 
     'HS-grad', 
     'Prof-school', 
     'Assoc-acdm', 
     'Assoc-voc', 
     '9th', 
     '7th-8th', 
     '12th', 
     'Masters', 
     '1st-4th', 
     '10th', 
     'Doctorate', 
     '5th-6th', 
     'Preschool'],
    ## marital.status
    ['Married-civ-spouse', 
     'Divorced', 
     'Never-married', 
     'Separated', 
     'Widowed', 
     'Married-spouse-absent', 
     'Married-AF-spouse'],
    # occupation
    ['Tech-support', 
     'Craft-repair', 
     'Other-service', 
     'Sales', 
     'Exec-managerial', 
     'Prof-specialty', 
     'Handlers-cleaners', 
     'Machine-op-inspct', 
     'Adm-clerical', 
     'Farming-fishing', 
     'Transport-moving', 
     'Priv-house-serv', 
     'Protective-serv', 
     'Armed-Forces'],
    # relationship
    ["Wife", 
     "Own-child", 
     "Husband", 
     "Not-in-family", 
     "Other-relative", 
     "Unmarried"],
    # race
    ["White", 
     "Asian-Pac-Islander", 
     "Amer-Indian-Eskimo", 
     "Other", 
     "Black"],
    # sex
    ["Female", 
     "Male"],
    # native
    ["United-States", 
     "Cambodia", 
     "England", 
     "Puerto-Rico", 
     "Canada", 
     "Germany", 
     "Outlying-US(Guam-USVI-etc)",
     "India", 
     "Japan", 
     "Greece", 
     "South", 
     "China", 
     "Cuba", 
     "Iran", 
     "Honduras", 
     "Philippines", 
     "Italy",
     "Poland", 
     "Jamaica", 
     "Vietnam", 
     "Mexico", 
     "Portugal", 
     "Ireland", 
     "France", 
     "Dominican-Republic", 
     "Laos",
     "Ecuador", 
     "Taiwan", 
     "Haiti", 
     "Columbia", 
     "Hungary", 
     "Guatemala", 
     "Nicaragua", 
     "Scotland", 
     "Thailand",
     "Yugoslavia", 
     "El-Salvador", 
     "Trinadad&Tobago", 
     "Peru", 
     "Hong", 
     "Holand-Netherlands"]
]
```

```python
col_transforms = [('cat_trans', preprocessing.OneHotEncoder(categories=categories),  ['workclass', 
                                                                 'education', 
                                                                 'marital.status', 
                                                                 'occupation', 
                                                                 'relationship', 
                                                                 'race', 
                                                                 'sex', 
                                                                 'native'])]
```

```python

## Compile the transformation
transformer = ColumnTransformer(transformers=col_transforms)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from catboost import CatBoostClassifier
import xgboost as xgb
```

```python
models = []

results = []

models.append(('DT',DecisionTreeClassifier()))
models.append(('SVM', SVC()))
models.append(('LR', LogisticRegression(solver='lbfgs', max_iter=1000)))
models.append(('RF', RandomForestClassifier(min_samples_split= 30, random_state=44)))
models.append(('KNN',KNeighborsClassifier()))
models.append(('XGB',xgb.XGBClassifier(objective="binary:logistic", random_state=48)))
models.append(('AB', AdaBoostClassifier()))
models.append(('EB', ExtraTreesClassifier(n_estimators=20, max_depth=None,min_samples_split=2, random_state=42)))
models.append(('GB',  GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=40)))
models.append(('CB',CatBoostClassifier(iterations=5,     learning_rate=0.1, random_state=44)))
models.append(('MLP',MLPClassifier(hidden_layer_sizes=(200,), activation='relu', solver="adam", max_iter=200, random_state =4)))
```

```python
for name, model in models:
   
    # print(names)
    pipe = Pipeline(steps=[('feat_trans', transformer), ('model',model)])
    pipe.fit(X_tr_inv_feats, X_tr_inv_labels)
    score = pipe.score(X_tr_inv_feats, X_tr_inv_labels)
    
    print('score',f'of the model:{name}........',score )
   
```

```python
model = MLPClassifier(hidden_layer_sizes=(200,), activation='relu', solver="adam", max_iter=200)

# define pipeline
pipeline_MLP = Pipeline(steps=[('feat_trans', transformer), ('model',model)]) 
pipeline_MLP.fit(X_tr_inv_feats, X_tr_inv_labels)
# Make soft prediction
pred_mlp = pipeline_MLP.predict_proba(X_ts_inv_feats)
print(f'Soft prediction: {pred_mlp}')
```

```python
model = DecisionTreeClassifier()

# define pipeline
pipeline_DT = Pipeline(steps=[('feat_trans', transformer), ('model',model)]) 
pipeline_DT.fit(X_tr_inv_feats, X_tr_inv_labels)
# Make soft prediction
pred_DT = pipeline_DT.predict_proba(X_ts_inv_feats)
print(f'Soft prediction: {pred_DT}')
```

```python
model = ExtraTreesClassifier(n_estimators=20, max_depth=None,min_samples_split=2, random_state=42)

# define pipeline
pipeline_ET = Pipeline(steps=[('feat_trans', transformer), ('model',model)]) 
pipeline_ET.fit(X_tr_inv_feats, X_tr_inv_labels)
# Make soft prediction
pred_ET = pipeline_ET.predict_proba(X_ts_inv_feats)
print(f'Soft prediction: {pred_ET}')
```
